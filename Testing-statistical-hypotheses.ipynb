* Testowanie hipotez: 

    Przeprowadzono weryfikace hipotez, dla dancyh numer $1$ i numer $2$. Dokładnie dla danych numer $1$ z rozkładu Normlanego $N(\mu,0.2)$, na poziomie ufności $\alpha = 0.05$ zweryfikowano hipotezy:

    * $\mu < 1.5$.

    * $\sigma^2 \neq 1.5$;
    * $\sigma^2 > 1.5$;
    * $\sigma^2 < 1.5$.
    
    W każdym z przypadków wyznaczono obszary krytyczne i p-wartości. Dodatkowo dla każdej ze zweryfikowanych hipotez dla danych $1$ i danych $2$ wyzanczono symulacyjnie prawdopodobieństwa popełnienia błedu $I$ rodzaju i błedu $II$ rodzaju. Ponadtow wyznaczono moce testów.

### 2.1 Zadanie 1: testowanie hipotez dla danych numer $1$ z rozkładu Normlanego $N(\mu,0.2)$, na poziomie ufności $\alpha = 0.05$:
   * $\mu \neq 1.5$

```python
#### H1: µ ̸= 1.5
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd
  
data = pd.read_csv("lista8_zad1.txt", header = None)

# Poziom istotności
alpha = 0.05

# Średnia próbki
sample_mean = np.mean(data)

# Odchylenie standardowe populacji
population_std = 0.2

# Liczba obserwacji w próbce
n = len(data)

# Wartość statystyki testowej
z = (sample_mean - 1.5) / (population_std / np.sqrt(n))
print(z)
# Obszary krytyczne (przyjmując rozkład normalny)
critical_region_left = stats.norm.ppf(alpha/2)
critical_region_right = stats.norm.ppf(1 - alpha/2)

p_value = 2 * (1 - stats.norm.cdf(abs(z)))

x = np.linspace(-7.5, 7.5, 100)
y = stats.norm.pdf(x, 0, 1)

# Wykres obszarów krytycznych
plt.plot(x, y)
plt.fill_between(x[x < critical_region_left], y[x < critical_region_left], color='red')
plt.fill_between(x[x > critical_region_right], y[x > critical_region_right], color='red')
plt.scatter(z,0,marker="s",color="g",label="Wartość statystyki testowej")
plt.title('Obszar krytyczny dla µ ̸= 1.5')
plt.xlabel('Wartość statystyki testowej H1')
plt.ylabel('Gęstość')
plt.legend(loc="upper left")
plt.show()

# Wyświetlenie wyników
#print('Wartość statystyki testowej (z-score):', z)
#print('Obszar krytyczny dla alternatywy H1: <', critical_region_left, ',', critical_region_right)
#print('p-wartość:', p_value)
```

* $\mu < 1.5$.

```python
#### H3: µ < 1.5
  
data = pd.read_csv("lista8_zad1.txt", header = None)

# Poziom istotności
alpha = 0.05

# Średnia próbki
sample_mean = np.mean(data)

# Odchylenie standardowe populacji
population_std = 0.2

# Liczba obserwacji w próbce
n = len(data)

# Wartość statystyki testowej
z = (sample_mean - 1.5) / (population_std / np.sqrt(n))

# Obszary krytyczne (przyjmując rozkład normalny)
critical_region_left = stats.norm.ppf(alpha)
p_value = 2 * (1 - stats.norm.cdf(abs(z)))

x = np.linspace(-7.5, 7.5, 100)
y = stats.norm.pdf(x, 0, 1)

# Wykres obszarów krytycznych
plt.plot(x, y)
plt.fill_between(x[x < critical_region_left], y[x < critical_region_left], color='red')
plt.scatter(z,0,marker="s",color="g",label="Wartość statystyki testowej")
plt.title('Obszar krytyczny dla µ < 1.5')
plt.xlabel('Wartość statystyki testowej H3')
plt.ylabel('Gęstość')
plt.legend(loc="upper left")
plt.show()

# Wyświetlenie wyników
print('Wartość statystyki testowej (z-score):', z)
print('Obszar krytyczny dla alternatywy H3 (lewostronny): <', critical_region_left)
print('p-wartość:', p_value)
```

### 2.1.1 Wnioski:
Jeśli obliczona przez nas wartość statystki testowej (zielony kwadrat) znajduje się w obszarze krytycznym (czerwone pole), to odrzucamy hipoteze zerową, przyjmując hipoteze alternatywną: 
- Podpunkt 1: $\mu \neq 1.5$, $H_0$ - odrzucamy;
- Podpunkt 2: $\mu > 1.5$, przyjmujemy - $H_0$;
- Podpunkt 3: $\mu < 1.5$, $H_0$- odrzucamy.

Następnie na mocy poniższej definicji, odpowiedzaino na pytanie co się stanie kiedy będziemy zminiać poziom ufności testu.

Definicja: relacja międyz poziomem istotności, a poziomem ufności:
   * $poziom~istotności=\alpha$;
   * $poziom~ufności=1-\alpha$.

Z powyższej definijci mamy proste wnioski:
- Kiedy zwiększamy $poziom~istotności=\alpha$, zmniejszamy $poziom~ufności=1-\alpha$;
- Kiedy zmniejszamy $poziom~istotności=\alpha$, zwiekszamy $poziom~ufności=a-\alpha$.

Zatem, kiedy zwiększymy $poziom~istotności=\alpha$, to zminiejszymy $poziom~ufności=1-\alpha$, a to powoduje powiększenie się obszaru krytycznego, czyli łatwiej nam odrzucić testowaną hipotezze. W sytuacji, gdy zmniejszymy $poziom~istotności=\alpha$, to zwiększymy $poziom~ufności=1-\alpha$, co skutkuje pomniejszeniem obszaru krytycznego, czyli trudniej nam odrzucić testowaną hipoteze.

### 2.2 Zadanie 2: testowanie hipotez dla danych numer $2$ z rozkładu Normlanego $N(0.2,\sigma^2)$, na poziomie ufności $\alpha = 0.05$:
   * $\sigma^2 \neq 1.5$

```python
#### H1: sigma^2 != 1.5

from scipy.stats import chi2

# Dane z próby
data = pd.read_csv("lista8_zad2.txt", header = None)

# Parametry rozkładu normalnego
mu = 0.2

# Wielkość próby
n = len(data)

# Obliczenie statystyki testowej
sample_variance = np.var(data, ddof=1)  # Ostrzeżenie: używamy nieobciążonej estymacji wariancji

# Obliczenie statystyki testowej chi-kwadrat
chi_squared = (n - 1) * sample_variance / 1.5

# Obliczenie p-wartości
p_value = 2 * min(chi2.cdf(chi_squared, df=n-1), 1 - chi2.cdf(chi_squared, df=n-1))

# Wyświetlenie wyników dla hipotezy sigma^2 != 1.5
#print("Hipoteza: sigma^2 != 1.5")
#print("Wartość statystyki testowej chi-kwadrat H1:", chi_squared)
#print("P-wartość:", p_value)

# Obszary krytyczne dla hipotezy sigma^2 != 1.5
alpha = 0.05

critical_region_left = chi2.ppf(alpha/2, df=n-1)
critical_region_right = chi2.ppf(1 - alpha/2, df=n-1)

#print("Obszar krytyczny (lewy): x <= {:.2f}".format(critical_region_left))
#print("Obszar krytyczny (prawy): x >= {:.2f}".format(critical_region_right))

df = n-1
x = np.linspace(800, 1200, 100)
density = chi2.pdf(x, df)

# Wykres obszarów krytycznych
plt.plot(x, density)
plt.fill_between(x[x < critical_region_left], density[x < critical_region_left], color='red')
plt.fill_between(x[x > critical_region_right], density[x > critical_region_right], color='red')
plt.scatter(chi_squared,0,marker="s",color="g",label="Wartość statystyki testowej")
plt.title('Obszar krytyczny dla sigma^2 != 1.5')
plt.xlabel('Wartość statystyki testowej H1')
plt.ylabel('Gęstość')
plt.legend(loc='upper left')
plt.show()
```
