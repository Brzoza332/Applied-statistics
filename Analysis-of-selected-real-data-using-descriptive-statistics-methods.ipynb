# 2. Podstawowe statystyki:
## 2.1 Wzory i definicje:
* Średnia arytmetyczna:
  $$\overline{x} = \frac{\sum_{i=1}^{n}x_{i}}{n};$$
* Średnia harmoniczna:
  $$\frac{n}{\sum_{i=1}^{n}\frac{1}{x_{i}}};$$
* Średnia ucinana:
 $$\frac{1}{n-2k}\sum_{i=k+1}^{n=k}x_{i};$$
* Średnia Winsorowska:
  $$\frac{1}{n}\left[(k+1)x_{k+1} + \sum_{i=k+2}^{n=k-1}x_{i+(k+1)} \cdot x_{(n-k)}\right];$$
* Mediana:
 kwantyl rzędu $\frac{1}{2};$

\begin{equation}
        x_{med} =
    \begin{cases} 
        x_{(n+1)/2} & \text{dla n nieparzystych;}\\
        \frac{1}{2}(x_{n/2}+x_{n/2+1}) & \text{dla n parzystych;}
    \end{cases}
 \end{equation}
* Kwartyle:

    Kwantyle to wartości cechy danej zbiorowości statystycznej, które umożliwiają jej podział na określone części w odpowiednich proporcjach. W praktyce do najczęściej stosowanych kwantyli zaliczamy: kwartyle, decyle i percentyle.
* Rozstęp międzykwartlowy:
  $$IQR = Q3 - Q1;$$
* Wariancja:
 $$S^2 = \frac{1}{n-1} \sum_{i=1}^{n}(x_{i}-\overline{x})^2 ;$$
* Odchylenie standardowe:
 $$S = \sqrt{S^2};$$
* Współczynnik zmienności:
 $$\vartheta = \frac{1}{\overline{x}\cdot 100%};$$
* Wspołczynnik skośności:
 $$\alpha = \frac{n}{(n-1)(n+2)}\sum_{i=1}^{n}\left(\frac{x_i-\overline{x}}{S}\right)^3;$$
* Kurtoza:
 $$K = \frac{n-1}{(n-2)(n-3)}((n+1)k-3(n-1))+3;$$
* Rozstęp:
 $$\text{rozstep} = \max(x_i) - \min(x_i).$$

## 2.2 Dane statystyczne dla wieku:

```python
from scipy.stats import mstats

# Obliczenie średniego wieku
mean_age = df['age'].mean()
print('Średni wiek: ', mean_age)

# Obliczenie średniej harmonicznej
harmonic_mean_age = len(df) / (1 / df['age']).sum()
print("Średnia harmoniczna: ", harmonic_mean_age)

# Obliczenie średniej ucinanej
n = int(0.1 * len(df)) # określenie liczby wartości ucinanych
sorted_values_age = sorted(df['age'])
trimmed_mean_age = np.mean(sorted_values_age[n:-n])
print("Średnia ucinana: ", trimmed_mean_age)

# Obliczenie średniej winsorowskiej
winsorized_mean_age = mstats.winsorize(df['age'], limits=[0.1, 0.1]).mean()
print("Średnia Winsorowska: ", winsorized_mean_age)

# Obliczenie mediany wieku
median_age = df['age'].median()
print('Mediana: ', median_age)

# Obliczenie odchylenia standardowego wieku
std_age = df['age'].std()
print('Odchylenie standardowe: ', std_age)

# Obliczenie współczynnika skośności
skewness_age = df['age'].skew()
print('Współczynnik skośności: ', skewness_age)

# Obliczenie maksymalnej wartości wieku
max_age = df['age'].max()
print('max: ', max_age)

# Obliczenie minimalnej wartości wieku
min_age = df['age'].min()
print('min: ', min_age,)

# Obliczenie współczynnika zmienności
cv_age = std_age / mean_age * 100
print('Współczynnik zmienności: ', cv_age)

# Obliczenie kurtozy
kurt_age = df['age'].kurtosis()
print('Kurtoza: ', kurt_age) #lekko ogonowe

# Obliczenie kwartyli
q1_age = df['age'].quantile(0.25)
print(f"Kwartyl rzędu 0.25: {q1_age}")
q3_age = df['age'].quantile(0.75)
print(f"Kwartyl rzędu 0.75: {q3_age}")

# Obliczenie IQR(rozstęp międzykwantylowy)
iqr_age = q3_age - q1_age
print("Rozstęp międzykwartylowy (IQR):", iqr_age)

# Obliczenie rozstępu
range_age = max_age - min_age
print("Rozstęp: ", range_age, '\n')
```

## 2.3 Interpetacja

#### Średnia wieku:
- Mówi nam o środku ciężkości zbioru danych, wnioskujemy, że większość osób zatrzymanych była w młodym wieku.

#### Średnia ucinana:
- Średnia ucinana jest zbliżona do średniej arytmetycznej, przez co wnioskować można, że skrajne wyniki nie wywierają dużego wpływu na wartość średniej arytmetycznej.

#### Mediana:
- Mediana jest mniejsza od wartości średniej, stąd zakładamy, że rozkład jest prawostronnie skośny.

#### Współczynnik skośności:
- Z dodatności współczynnika skośności potwierdzamy prawostronną skośność (asymetrie dodatnią).

#### Współczynnik zmienności:
- Wynik około 35% mówi nam o przeciętnym zróżnicowaniu danych.

#### Kurtoza:
- Kurtoza większa od zera oznacza, że rozkład jest bardziej "szpiczasty" niż rozkład normalny (bardziej skupiony wokół średniej) i ma dłuższe "ogony" (więcej obserwacji znajduje się w częściach rozkładu odchodzących daleko od średniej). Taki rozkład nazywa się leptokurtycznym.

#### Rozstęp miedzykwartylowy:
- Wskazuje na przeciętne zróżnicowanie danych.

## 3.3 Gęstość empiryczna:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math 
# Wczytanie danych z pliku CSV
data = pd.read_csv('Arrests.csv')


x = []
y = []
z = []

with open('Arrests.csv') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        x.append(row['age'])
        y.append(row['year'])
        z.append(row['checks'])
        
x_sort = sorted(x)
y_sort = sorted(y)
z_sort = sorted(z)


# Obliczenie dystrybuanty empirycznej
n = len(x_sort)
x_lin = np.arange(1, n + 1) / n

n1 = len(y_sort)
y_lin = np.arange(1, n1 + 1) / n1

n2 = len(z_sort)
z_lin = np.arange(1, n2 + 1) / n2

# Wyświetlenie wykresu dystrybuanty empirycznej
lambd = 17.5
X=np.arange(8, 66)
def distribution(X, num):
    X = np.array([X.tolist()[0]-1]+X.tolist()+[X.tolist()[-1]+1])
    Y = np.array([sum(num[:i+1]) for i in range(len(num))])
    Y = np.insert(Y, [0, 0], 0)
    return X, Y

def prawdoter(t):
    t=t.tolist()
    for i in range(len(t)):
        t[i]=(lambd**i)*math.e**(-lambd)/math.factorial(i)
    return t

Xt2, Yt2 = distribution(X,prawdoter(X))

remove_idx = [0,1,2,3,4,5,6]
Xt2 = np.delete(Xt2, remove_idx)
Yt2 = np.delete(Yt2, remove_idx)
plt.figure(figsize=(14,6))
plt.plot(x_sort, x_lin,'purple',label='Empirical distribution')
plt.xlabel('Age')
plt.ylabel('Empirical distribution')
plt.legend()
plt.show()

plt.figure(figsize=(14,6))
plt.plot(y_sort, y_lin,'purple')
plt.xlabel('year')
plt.ylabel('Empirical distribution')
plt.show()

plt.figure(figsize=(14,6))
plt.plot(z_sort, z_lin,'purple')
plt.xlabel('checks')
plt.ylabel('Empirical distribution')
plt.show()
```

## 4.1 Dystrybuanta teoretyczna i gęstość teoretyczna

#### Dopasowano do gęstości i dystrybuanty empirycznej (pierwszej badanej tezy), znany najbardziej zbliżony rozkład.
